```python
print("=== Основні типи даних ===")

a = "змінна з текстом"
b = 1                 # int
b1 = 1.1              # float
c = ["a", 1, 1.25, "Слово", a]   # list
d = {"a": "Слово", "b": 1, a: b} # dict
e = ("a", a)          # tuple
f = {"ss", str(b)}    # set

print("a:", a)
print("b:", b)
print("b1:", b1)
print("c (list):", c)
print("d (dict):", d)
print("e (tuple):", e)
print("f (set):", f)
```

```python
print("\n=== Вбудовані константи ===")
print("True:", True)
print("False:", False)
print("None:", None)

print(f"Форматований вивід працює ось так: {True}")

import sys
print("\n=== Зарезервовані слова Python ===")
help("keywords")
```

```python
print("\n=== Вбудовані функції ===")

print("abs(-12.5) =", abs(-12.5))
print("len('Python') =", len("Python"))
print("type(123) =", type(123))

print(
    abs(-12.5),
    f"є рівним {abs(12.5)}",
    "і якщо порівняти то:",
    abs(-12.5) == abs(12.5)
)
```

```python
print("\n=== Цикли ===")

letters = ["a", "b", "c"]

for i in range(len(letters)):
    print(f"На позиції {i} знаходиться буква {letters[i]}")
else:
    print("Ця конструкція безглузда!")

print("\nЦикл while:")
i = 0
while i < 3:
    print("i =", i)
    i += 1
```

```python
print("\n=== Розгалуження ===")

from random import randint

A = randint(0, 1)
if A == 1:
    print(f"Значить A = {A}")
else:
    print("Але може бути що A = {}".format(A))
```

```python
print("\n=== Обробка помилок ===")

A = 0
try:
    print("Що буде якщо", 10 / A, "?")
except Exception as e:
    print("Невже це помилка >", e)
finally:
    print("О це так на тобі!")
```

```python
print("\n=== Контекст-менеджер with ===")

with open("README.md", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        print(f"{i})> {line.strip()}")
```

```python
print("\n=== Lambda-функції ===")

def a_b_func(a, b):
    return a, b

this_is_lambda = lambda first, age: f"Цей код написав: {first}, Мені {age} років"

print("Це просто функція:", a_b_func)
print("А це лямбда:", this_is_lambda)
print("Це її виклик:", this_is_lambda("Аліна", 17))
print(this_is_lambda(*a_b_func("Python", 1)))
```
