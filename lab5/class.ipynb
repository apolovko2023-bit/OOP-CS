Методи класу

```python
# Методи класу: застосування до всіх об'єктів a1-a3

class MyAnimals:
    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals}, яку звати {name}, їй {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass

    def speak(self):
        if self.animals == "Кіт":
            return f"Моя тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя тваринка {self.animals} говорить Гавв"
        else:
            return f"Моя тваринка {self.animals} не розмовляє"

    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила зі мною ще один рік"


# Створення об'єктів
a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals(animals="Собака", name="Рекс", age=2)

print("-" * 50)

# Застосування методів до ВСІХ об'єктів
animals_list = [a1, a2, a3]

for animal in animals_list:
    print(f"Тваринка {animal.animals}:")
    print(animal.speak())
    print(f"Вік до: {animal.age}")
    print(animal.add_one_year())
    print(f"Вік після: {animal.age}")
    print("-" * 50)
```

Змінні класу, глобальні змінні над всіма обєктами

```python
# Змінні класу (глобальні для всіх об'єктів)

class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals}, яку звати {name}, їй {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass

        # Робота зі змінними класу
        MyAnimals.total_animals += 1
        MyAnimals.total_animal_types.add(animals)


# Початковий стан змінних класу
print(
    f"На початку: типи тварин {list(MyAnimals.total_animal_types)}, "
    f"кількість тварин {MyAnimals.total_animals}"
)

print("-" * 50)

# Дані для створення об'єктів у циклі
animals_data = [
    ("Рибки", "Неони", 1),
    ("Рибки", "Гупії", 1),
    ("Рибки", "Соміки", 1),
    ("Кіт", "Мурка", 5, 10),
]

objects = []

# Створення об'єктів у циклі
for data in animals_data:
    obj = MyAnimals(*data)
    objects.append(obj)

print("-" * 50)

# Стан змінних класу після створення об'єктів
print(
    f"Після створення об'єктів: типи тварин {list(MyAnimals.total_animal_types)}, "
    f"кількість тварин {MyAnimals.total_animals}"
)

print("-" * 50)

# Доступ до змінної класу через об'єкт
a1 = objects[0]
print(f"Доступ до змінної класу через об'єкт: {a1.total_animal_types}")

# Перезапис змінної класу в об'єкті
a1.total_animal_types = {"Собака"}

print(
    f"Після зміни через об'єкт:\n"
    f" - через об'єкт: {a1.total_animal_types}\n"
    f" - через клас: {MyAnimals.total_animal_types}"
)
```

Статичні методи та змінна self

```python
# Статичні методи та змінна self

class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила зі мною ще один рік"

    @staticmethod
    def call_pet():
        return "Тваринка біжить до нас"


# Створення об'єктів
a1 = MyAnimals("Рибки", "Гупії", 1)
a2 = MyAnimals("Кіт", "Мурка", 5)

animals_list = [a1, a2]

# Виклик статичного методу для кожного об'єкта
for animal in animals_list:
    print(f"{animal.animals}: {animal.call_pet()}")

print("-" * 50)

# Виклик статичного методу з класу
print(f"Виклик зі сторони класу: {MyAnimals.call_pet()}")

print("-" * 50)

# Демонстрація роботи self
print(f"Виклик методу з об'єкта: {a1.add_one_year()}")
print(f"Виклик методу з класу: {MyAnimals.add_one_year(a1)}")
```

Робота з properties

```python
from datetime import datetime

# Робота з properties
class MyAnimals:
    max_age = 20

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    @property
    def live_untill(self):
        """Рік, до якого тваринка може жити"""
        return datetime.today().year - self.age + self.max_age

    @property
    def info(self):
        """Додаткова property з інформацією про тваринку"""
        return f"Тваринка: {self.animals}, Ім'я: {self.name}, Вік: {self.age}"

    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила зі мною ще один рік"


# Створення об'єкта
a1 = MyAnimals("Кіт", "Мурка", 5)

# Вивід properties
print(a1.info)
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")

# Зміна віку
a1.add_one_year()

# Повторний вивід properties
print(a1.info)
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
```

Методи Класу

```python
# Методи класу: перевірка імені з великої літери

class MyAnimals:
    total_csv = 0
    total_objects = 0

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = MyAnimals.check_name(name)
        self.age = int(age)
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data: str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)

    @classmethod
    def check_name(cls, name: str) -> str:
        """
        Перевіряє, щоб ім'я починалося з великої літери
        """
        if not name:
            return name
        return name.capitalize()


# Створення об'єктів
a1 = MyAnimals.from_csv("Кіт,мурка,5")
a2 = MyAnimals("Собака", "рекс", 2)

# Вивід результатів
print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")

print(
    f"Загальна кількість об'єктів {MyAnimals.total_objects}, "
    f"створених з CSV формату {MyAnimals.total_csv}"
)
```

Магічні методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        return f"Це є обєкт {self.__repr__()}"

a1 = MyAnimals("Собака", "Рекс", 2)
print(a1)

print(repr(a1))

str(a1)
```
